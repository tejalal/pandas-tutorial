# 311 Service request dataset
- We're going to use a new dataset here, to demonstrate how to deal with larger datasets. 
- This is a subset of the of 311 service requests from [NYC Open Data](https://nycopendata.socrata.com/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9). 

```python
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
complaints = pd.read_csv('311-service-requests.csv')
complaints.shape # rows, columns
```

# Summary

```python
complaints
```

# Selecting columns and rows

```python
complaints['Descriptor']
```

```python
complaints[:3]
```

We can combine these to get the first 3 rows of a column:

```python
complaints['Complaint Type'][:3]
```

# Selecting multiple columns

```python
complaints[['Complaint Type', 'Descriptor']][0:3]
```

# Finding most common complaint type?

```python
complaints['Complaint Type'].value_counts()
```

### Top 5 most common complaints

```python
complaint_counts = complaints['Complaint Type'].value_counts()
complaint_counts[:5]
```

```python
complaint_counts[:5].plot(kind='pie')
```

### Which borough has the most noise complaints

```python
noise_complaints = complaints[complaints['Complaint Type'] == "Noise - Street/Sidewalk"]
noise_complaints[:5]
```

### Combine more than one condition with the `&` operator

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
in_brooklyn = complaints['Borough'] == "BROOKLYN"
complaints[is_noise & in_brooklyn][:5]
```

If we just wanted a few columns:

```python
complaints[is_noise & in_brooklyn][['Complaint Type', 'Borough', 
                                    'Created Date', 'Descriptor']][:10]
```

# So, which borough has the most noise complaints?

```python
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
noise_complaints = complaints[is_noise]
noise_complaints['Borough'].value_counts()
```

It's Manhattan! But what if we wanted to divide by the total number of complaints, to make it make a bit more sense?

```python
noise_complaint_counts = noise_complaints['Borough'].value_counts()
complaint_counts = complaints['Borough'].value_counts()
```

```python
noise_complaint_counts / complaint_counts.astype(float)
```

```python
(noise_complaint_counts / complaint_counts.astype(float)).plot(kind='bar')
```

So Manhattan really does complain more about noise than the other boroughs! Neat.
