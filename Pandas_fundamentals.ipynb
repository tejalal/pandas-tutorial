## Workinng with GroupBy and Aggregation

```python
data.groupby('class').size()
```

```python
data.groupby(['class','sepal-length']).groups
```

```python
data.groupby('class').mean()
```

```python
# Iterating through Groups
grouped = data.groupby('class')
for cls,values in grouped:
    print(cls)
    print(values)
```

```python
# Select a Group

grouped = data.groupby('class')
grouped.get_group('Iris-setosa')
```

## Aggregations
- An aggregated function returns a single aggregated value for each group. 
- Once the group by object is created, several aggregation operations can be performed on the grouped data.

```python
grouped = data.groupby('class')
grouped['sepal-length'].agg(np.mean)
```

```python
# Mutiple aggregate functions at once

grouped = data.groupby('class')
grouped['sepal-length'].agg([np.sum, np.mean, np.std])
```

## Working with Text data

- Pandas has a set of string functions, which make it easy to operate on string data. 
- Most importantly, these functions ignore (or exclude) missing/NaN values.
- Almost, all of these methods work with Python string functions. 
- So, convert the Series Object to String Object and then perform the operation.

- upper()
- lower()
- title()
- len()
- isnumeric()
- strip()
- isdigit()
- isalpha()
- isupper()
- islowe()
- swapcase()
- count(value)
- find(value)
- findall(value)
- startswith()
- endswith()
- split(seprator)
- replace(value1,value2)
- cat(seperator)
- repeat(n)
- contains(value)
- get_dummies()

```python
countries = pd.Series(['India', 'australia', ' USA', 'China'])

countries = countries.str
print(countries)
```

```python
print (countries.upper()) 
print (countries.lower())
print (countries.title()) 
print (countries.len()) 
print (countries.strip()) 
print (countries.isnumeric()) 
print (countries.isdigit()) 
print (countries.isalpha()) 
```

```python
print (countries.isupper()) 
print (countries.islower()) 
print (countries.swapcase()) 
print (countries.count('U')) 
print (countries.find('USA')) 
print (countries.findall('a')) 

print (countries.findall('a')) 
```

```python
print(countries.startswith('a')) 
print(countries.endswith('a')) 
print(countries.split(',')) 

print(countries.replace('a', 'b')) 

print(countries.cat(sep='$')) 

print(countries.contains('aus')) 

print(countries.get_dummies()) 

print(countries.repeat(2)) 
```
